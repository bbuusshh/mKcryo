```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from scipy import constants as cc
```

```python
temp = np.array([0, 5, 10, 15])
lam = np.array([621.55, 622.56, 623.59, 624.55])
```

```python
coef = np.polyfit(temp,lam,1)
poly1d_fn = np.poly1d(coef) 
```

```python

plt.figure(figsize=(10,7))
plt.plot(temp, poly1d_fn(temp))
temp2 = np.linspace(-20, 20, 10)
plt.plot(temp2, poly1d_fn(temp2))
plt.scatter(temp, lam)

plt.xlabel("Temp, oC")
plt.ylabel("lambda, nm")
```
